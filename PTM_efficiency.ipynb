```python
import matplotlib.pyplot as plt
import numpy as np

# Data from the table
years = [2022, 2023, 2024, 2025]
batches = [20, 2, 21, 9]  # Total No. of batches for PTM
quantities = [2977.94, 339.62, 5997.09, 2843.04]  # Total Qty produced for PTM

# Calculate efficiency (Average Qty per Batch)
efficiency = np.array(quantities) / np.array(batches)

# Create a bar chart
plt.figure(figsize=(8, 5))
plt.bar(years, efficiency, color=['blue', 'orange', 'green', 'red'])

# Labels and title
plt.xlabel("Year")
plt.ylabel("Average Quantity per Batch")
plt.title("PTM Efficiency Over the Years")
plt.xticks(years)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the values on bars
for i, v in enumerate(efficiency):
    plt.text(years[i], v + 50, f"{v:.2f}", ha='center', fontsize=10)

# Show the plot
plt.show()
```
